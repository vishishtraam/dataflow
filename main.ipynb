```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
product_trend = sales.groupby([sales['Invoice Date'].dt.month, 'Product'])['Units Sold'].sum().unstack()

plt.figure(figsize=(12,6))
product_trend.plot(kind='line', marker='o')
plt.title('Product Demand by Month')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(title='Product')
plt.show()
```

```python
method_sales = sales.groupby('Sales Method')['Total Sales'].sum()

plt.figure(figsize=(6,5))
method_sales.plot(kind='bar', color='orange')
plt.title('Sales by Method')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = sales.groupby('Region')['Total Sales'].sum()

plt.figure(figsize=(6,6))
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Region')
plt.ylabel('')
plt.show()
```

```python
retailer_sales = sales.groupby('Retailer')['Total Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
retailer_sales.plot(kind='bar', color='skyblue')
plt.title('Top Retailers by Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
